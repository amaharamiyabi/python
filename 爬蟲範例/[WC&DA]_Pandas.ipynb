```python
#dataframe at和iat取值/更新值
cities = {'country':['China', 'Japan', 'Singapore'],
      'town':['Beijing','Tokyo','Singapore'],
      'population':[2000, 1600, 600]}
city_df = pd.DataFrame(cities, index=[1,2,3])
print(city_df)

print('-'*50)
at_df = city_df.at[1,'country']
print(at_df)

print('-'*50)
city_df.at[1,'country'] = 'Taiwan'
print(city_df)

print('-'*50)
iat_df = city_df.iat[0,1]
print(iat_df)

print('-'*50)
city_df.iat[0,1] = 'Taipei'
print(city_df)
```

```python
#dataframe loc取值
cities = {'country':['China', 'Japan', 'Singapore'],
      'town':['Beijing','Tokyo','Singapore'],
      'population':[2000, 1600, 600]}
city_df = pd.DataFrame(cities, index=[1,2,3])
print(city_df)
print('-'*50)
#返回符合index=1的整筆資料
loc_df=city_df.loc[1]
print(loc_df)
print('-'*50)
#返回符合index=1且columns='town'的資料
loc_df=city_df.loc[1]['town']
print(loc_df)
print('-'*50)
#返回全部index且columns='town'的資料
loc_df=city_df.loc[:]['town']
print(loc_df)
print('-'*50)
#返回index=[1,3]且columns=['town','population']的資料
loc_df=city_df.loc[[1,3],['town','population']]
print(loc_df)
print('-'*50)

```

```python
#dataframe iloc取值
cities = {'country':['China', 'Japan', 'Singapore'],
      'town':['Beijing','Tokyo','Singapore'],
      'population':[2000, 1600, 600]}
city_df = pd.DataFrame(cities, index=[1,2,3])
print(city_df)
print('-'*50)
#返回符合index編號為1的整筆資料
loc_df=city_df.iloc[1]
print(loc_df)
print('-'*50)
#返回符合index編號為1且columns編號為1的資料
loc_df=city_df.iloc[1][1]
print(loc_df)
print('-'*50)
#返回全部index編號且全部columns編號的資料
loc_df=city_df.iloc[:][1:2]
print(loc_df)
print('-'*50)
#返回index編號為1到2的資料
loc_df=city_df.iloc[[1,2]]
print(loc_df)
print('-'*50)
```

```python
#dataframe合併
cities = {'country':['China', 'Japan', 'Singapore'],
      'town':['Beijing','Tokyo','Singapore'],
      'population':[2000, 1600, 600]}
city_df = pd.DataFrame(cities, index=[1,2,3])
city_new = {'country':'France', 'town':'Paris'}
city_new_df = pd.DataFrame(city_new, index=[1])

city_concat=pd.concat([city_df,city_new_df], ignore_index=True)
print(city_concat)
print('-'*50)

city_concat=pd.concat([city_df,city_new_df])
print(city_concat)
print('-'*50)

city_new_2 = {'countrys':'France', 'town':'Paris', 'population':1800}
city_new_df2 = pd.DataFrame(city_new_2, index=[1])
city_concat=pd.concat([city_df,city_new_df2], ignore_index=True)
print(city_concat)
print('-'*50)

city_concat=pd.concat([city_df,city_new_df2], ignore_index=True, join='inner')
print(city_concat)
print('-'*50)

cities_2 = {'country':['China', 'Japan', 'Singapore'],
      'language':['Mandarin', 'Japanese', 'English']}
city_df2 = pd.DataFrame(cities_2, index=[1,2,3])
print(city_df)
print(city_df2)
city_merge=pd.merge(city_df, city_df2, on='country')
print(city_merge)

```
