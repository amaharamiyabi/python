```python
#user-agent
import requests

url = "http://httpbin.org/user-agent"

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
r = requests.get(url, headers=headers)
print(r.text)
```

```python
#BeautifulSoup soup()
import requests
from bs4 import BeautifulSoup

url = "https://fchart.github.io/"
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "lxml")
    #存取第一個某種標籤
    tag_a = soup.a
    tag_b = tag_a.find('b')
    print(tag_a)
    print(tag_b)
    print("-"*300)
    #存取所有某種標籤
    tags = soup("a")
    print(tags)
    print("-"*300)
    #存取tag內某個屬性的值
    for tag in tags:
        print(tag.get("href", None))

else:
    print("錯誤! HTTP請求失敗...")
```

```python
#BeautifulSoup find find_all _class
import requests
from bs4 import BeautifulSoup

url = "https://fchart.github.io/Elements.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
#存取某一個標籤的第一項tag節點
tag = soup.find("h2")
print("h2: ", tag.text)
tag = soup.find("b")
print("b: ", tag.text)
print('-'*100)

#存取全部符合指定標籤的tag節點
tags = soup.find_all("b")
print("b: ", tags[0].text)
print("b: ", tags[1].text)
print("b: ", tags[2].text)
print('-'*100)

#存取某一個標籤且某一個屬性=指定值的tag節點
#tag = soup.find("li", {"id":"q2"})
tag = soup.find("li", id='q2')
tag_q = tag.find("b")
print("Question: ", tag_q.text)
print('-'*100)

#存取全部符合指定標籤且某一個屬性=指定值的tag節點
tags_a = tag.find_all("li", {"class":"response"})
for tag in tags_a:
    print("Ans: ", tag.text)
print('-'*100)

#存取某一個標籤且指定類別選擇器的tag節點
tag_ans1 = soup.find("li", class_="response")
print(tag_ans1.text)
```

```python
#BeautifulSoup tag name get text attrs
import requests
from bs4 import BeautifulSoup

url = "https://fchart.github.io/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
tags = soup("a")
tag = tags[12]
print(tag)
print("標籤名稱: ", tag.name)
print("URL網址: ", tag.get("href", None))
print("標籤內容: ", tag.text)
print("target屬性: ", tag["target"])

#更新屬性值
tag["target"]="new_blank"
print("target屬性: ", tag["target"])
tag=soup.find('a', target="new_blank")
print(tag)

#刪除某個屬性
del tag["target"]
#print("target屬性: ", tag["target"])

print("-"*100)
tags = soup("img")
tag = tags[1]
print("圖片網址: ", tag.get("src", None))
print("alt屬性: ", tag["alt"])
print("屬性: ", tag.attrs)
print("多值屬性: ", tag['class'])
```

```python
#BeautifulSoup contents children descendants
import requests
from bs4 import BeautifulSoup

url = "https://fchart.github.io/Elements.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
tag = soup.find("body")
print("body的子節點 ", tag.contents)
print("-"*100)

print("-"*100)

for child in tag.children:
  print("body的子節點 ", child)
  #print("child的string ", child.string)
print("-"*100)
for des in tag.descendants:
  print("body的子節點 ", des)
```

```python
#BeautifulSoup strings
import requests
from bs4 import BeautifulSoup

url = "https://fchart.github.io/Elements.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
tag = soup.find("body")
print(tag.strings)
for s in tag.strings:
  print("當前標籤裡面的內容 :", s)

print("-"*100)
for s in tag.stripped_strings:
  print("當前標籤裡面的內容 :", s)
```

```python
#BeautifulSoup parents brother
import requests
from bs4 import BeautifulSoup

url = "https://fchart.github.io/Elements.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
tag = soup.find("body")
print("body的父節點 ", tag.parent)

print("-"*100)

tags = soup.find("li")
for parent in tags.parents:
  print("li的父節點 ", parent)

print("-"*100)
print("li的下一個兄弟節點 ", tags.next_sibling)
print("li的上一個兄弟節點 ", tag.previous_sibling)

print("-"*100)
for bro in tags.next_siblings:
  print("li的下一個兄弟節點 ", bro)
for bro in tag.previous_sibling:
  print("li的上一個兄弟節點 ", bro)
```
